```python
#ch='chr21'; gene='ENSG00000160213.5' ; var_str=45196326 ; var_snp=45214684

import statsmodels.formula.api as smf
import statsmodels.api as sm
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

DATA = "/storage/szfeupe/Runs/GTEx_estr/Analysis_by_Tissue/SUMMARY/Merged_Causality_best_tissue.table"
COLORS = ['b', 'r', 'g', 'm', 'y', 'k', 'grey']
BASEDIR = "/storage/szfeupe/Runs/GTEx_estr/Analysis_by_Tissue/"
SNPS = "/storage/szfeupe/Runs/GTEx_estr/SNP_Analysis"
STRS = "/storage/szfeupe/Runs/GTEx_estr/Normalized_Genotypes/"

Gene_table = pd.read_csv('/storage/szfeupe/Runs/GTEx_estr/FEATURES/Genes_only_table', sep='\t')
Data = pd.read_csv("/storage/szfeupe/Runs/GTEx_estr/Analysis_by_Tissue/WholeBlood/Master.table", sep='\t')



Data = pd.read_csv(DATA, sep='\t')
data = Data.sort_values('best.score', ascending=[False])
data0 = data.head(5)
examples = Data.loc[Data['gene'].isin(['ENSG00000117151.8', 'ENSG00000100852.8', 'ENSG00000090857.9'])]
data = pd.concat([data0,examples])

chs = list(data['chrom'])
genes = list(data['gene'])
starts = list(data['best.str.start'])
tissues = list(data['best.tissue'])
scores = list(data['best.score'])

for i in range(len(scores)):
    try: 
        print ('\n....',i,'....',str(tissues[i]))
    except:
        print ('PASS************')
        continue
    print('------------->',chs[i],' gene ',genes[i],'Causlity Score: ',scores[i],'\n')
    snps=pd.read_csv(BASEDIR+tissues[i]+"/SNP_Analysis/PQValues.txt", sep="\t")
    expr = pd.read_csv(BASEDIR+tissues[i]+"/Corr_Expr.csv",sep=',' ) 
    snp =pd.read_csv(SNPS+"/SNP_Geno"+chs[i], sep='\t') #snps
    Genos=pd.read_csv(STRS+"STR_Raw_Gen."+chs[i], sep='\t')
    
    S = snps.loc[snps['gene']==genes[i]]['str.start']; S=int(S)
    Gene = Gene_table.loc[Gene_table['gene']==genes[i]]['gene.name'].values[0]
    
    samples = list(expr.index)
    
    Genosnp = snp.loc[snp['start']==S,samples]
    #print(Genosnp.shape)
    Genostr = Genos.loc[Genos['start']==starts[i], samples]
    Expr = expr[genes[i]]
    
    Genosnp = Genosnp.replace('None', np.nan) ; #print(Genosnp)
    Genostr = Genostr.replace('None', np.nan) ; #print(Genostr)
    #samples = list(expr.index)
    
    plt.figure(figsize=(11,3))
    plt.subplot(131)
    plt.title(tissues[i]+' tissue')
    plt.scatter(Genosnp, Expr, label='Lead SNP', color='b',marker='.')
    plt.ylabel(Gene+' expression')
    plt.xlabel('Best SNP genotypes')
    plt.subplot(132)
    plt.scatter(Genostr, Expr,label='eSTR', color='g',marker='.')
    #plt.ylabel(Gene+' Gene expression')
    plt.xlabel('eSTR genotypes \nSTR_'+str(int(starts[i])))
 
    ##Hold snp genotypes constant
    geno=list(set(Genosnp.values.tolist()[0]))
    geno=list(set([x for x in np.nan_to_num(geno) if x!='nan']))
    n=len(geno) ; N=0
    ind=Genosnp.index[0]
    #fig = plt.figure(figsize=(8,6))
    plt.subplot(133) #
    for i in range(n):
        S = Genosnp.T.loc[Genosnp.T[ind]==geno[i]]
        sample=list(S.index)
        S['Exp']=Expr.loc[sample]
        S['Str']=Genostr.T.loc[sample][Genostr.index[0]]
        #print(N, list(S['Str']))
        S['Str'] = S['Str'].astype(float)
        S['Str']+=N
        if len(S['Exp'])==0:
            continue
        #ax = fig.add_subplot(n,1,1+i)
        print(geno[i], S.shape)
        plt.scatter(list(S['Str']), list(S['Exp']), color=COLORS[i], marker='.')
#        plt.title.set_text(Gene+' Expression @ SNP_genotype='+str(geno[i])+' ( n='+str(len(S['Exp']))+' )')
        if max(list(S['Str'].values))-min(list(S['Str'].values))!=0:
            N += max(list(S['Str'].values))-min(list(S['Str'].values))+10
        else:
            N +=max(list(S['Str'].values))+10
        #print(N, list(S['Str']))
    plt.xlabel('STRs at SNPs genotypes \nSTR_'+str(int(starts[i])))
    plt.xticks([])
    plt.show()    
```
