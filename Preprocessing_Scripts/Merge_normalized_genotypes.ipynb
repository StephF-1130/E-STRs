```python
#!/usr/bin/env python

import pandas as pd
import subprocess
import sys
import os


"""
    Merge Genotype files into ine single file 
    Output is a vcf file to be indexed with tabix
"""
Original = '/storage/szfeupe/Runs/650GTEx_estr/Genotypes/STR_Norm_lized_Geno.table'
Homopoly = '/storage/szfeupe/Runs/650GTEx_estr/Genotypes/Normalized_Homopolymers.table'
AddHET01 = '/storage/szfeupe/Runs/650GTEx_estr/Genotypes/Normalized_Additional_Het01.table'


S1 = pd.read_csv(Original, sep='\t')  #Original had HET(0.3) without homopolymers
S2 = pd.read_csv(Homopoly, sep='\t')  #We did Homopolymers only to add to the set
S3 = pd.read_csv(AddHET01, sep='\t')  #We moved HET threshold to 0.1 (This set only has 0.1<=HET<0.3)

AllGeno = pd.concat([S1,S2,S3])
AllGeno = AllGeno.sort_values(['chrom','start'])

AllGeno
```

```python
AllGeno.to_csv('/storage/szfeupe/Runs/650GTEx_estr/Genotypes/GenotypesNormalized',sep='\t', index=None)
print('end')
```
